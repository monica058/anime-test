### Import modules

```python
import pandas as pd
import datetime as dt
import numpy as np
```

### File Paths

```python
# Inpath
df_anime = pd.read_csv('anime.csv')
```

```python
# Outpath
outpath = r'D:\Documents\Projects\Anime\Output\Anime Data.xlsx'
```

### Data manipulation

```python
print(df_anime.shape)
df_anime.head()
```

```python
# Create ID
df_anime = df_anime.reset_index()
df_anime = df_anime.rename(columns={'index':'id'})
df_anime['studios'] = df_anime['studios'].str.replace("\['","").str.replace("\']","")
df_anime.head()
```

```python
df_anime['tags']
```

```python
df_anime.dtypes
```

```python
# Create tags dataframe
df_tags = df_anime[['id','tags']]
#df_tags['tags'] = df_tags['tags'].str.replace("\s+","")
df_tags['tags'] = df_tags['tags'].str.replace("[","").str.replace("]","").str.replace("'","")
df_tags['tags_list'] = df_tags['tags'].str.split(", ")
df_tags = df_tags[['id','tags_list']].explode('tags_list')
df_tags['tags_list'] = df_tags['tags_list'].str.strip()
df_tags = df_tags[df_tags['tags_list'] != ""]
df_tags.head()
```

```python
# Create contentwarn dataframe
df_warn = df_anime[['id','contentWarn']]
df_warn['contentWarn'] = df_warn['contentWarn'].str.replace("[","").str.replace("]","").str.replace("'","")
df_warn['contentWarn_list'] = df_warn['contentWarn'].str.split(", ")
df_warn = df_warn[['id','contentWarn_list']].explode('contentWarn_list')
df_warn['contentWarn_list'] = df_warn['contentWarn_list'].str.replace("\s+","")
df_warn['contentWarn_list'] = np.where(df_warn['contentWarn_list'] == "","Safe for Viewing",df_warn['contentWarn_list'])
print(df_warn.shape)
df_warn.head()
```

```python
# Create studio dataframe
df_studio = df_anime[['id','studios']]
df_studio['studios'] = df_studio['studios'].str.replace("[","").str.replace("]","").str.replace("'","")
df_studio['studios_list'] = df_studio['studios'].str.split(", ")
df_studio = df_studio[['id','studios_list']].explode('studios_list')
df_studio['studios_list'] = df_studio['studios_list'].str.replace("\s+","")
df_studio = df_studio[df_studio['studios_list'] != ""]
print(df_studio.shape)
df_studio.head()
```

```python
writer = pd.ExcelWriter(outpath, engine='xlsxwriter')
df_anime.drop(['contentWarn','tags','studios'], axis=1).to_excel(writer, index=False, sheet_name='Anime')
df_tags.to_excel(writer, index=False, sheet_name='Genre')
df_warn.to_excel(writer, index=False, sheet_name='ContentWarning')
df_studio.to_excel(writer, index=False, sheet_name='Studio')
writer.save()
writer.close()
```
